```python
# 각 종별로 가장 큰 값과 가장 작은 값의 비율을 구한다면?
# agg()
def get_ratio(x) :
    return x.max() / x.min()
```

```python
iris_species = iris.groupby(iris['species'])
```

```python
for key, group in iris_species :
    print('key : ', key)
    print('number : ', len(group))
    print(group)
    print('\n')
```

```python
iris_species.agg(get_ratio)
```

```python
# 종별로 꽃잎길이가 가장 긴 3개의 데이터를 뽑아낸다면?

def max3_petal(df) :
    return df.sort_values(by='petal_length', ascending=False)[:3]
```

```python
iris.groupby(iris['species'])
```

```python
iris.groupby(iris.species).apply(max3_petal)
```

- transform() : 그룹별 계산은 동일하지만 데이터프레임 자체를 변환
- 꽃잎길이가 해당 종 내에서 대/중/소 어느것에 해당되는지에 대한 프레임을 만들고 싶다면?
- cut() : 동일 길이로 나누어서 범주를 만들고 그룹에 대한 통계량 제공
- qcut() : 동일 갯수로 나누어서 범주를 만들고 그룹에 대한 통계량 제공 

```python
iris.groupby(iris['species'])
```

```python
iris.groupby(iris.species).petal_length.size()
```

```python
def cate3_petal_length(x) :
    return pd.qcut(x, 3 , labels=['소','중','대']).astype(str)
```

```python
iris['category'] =iris.groupby(iris.species).petal_length.transform(cate3_petal_length)
iris.head()
```

```python
# 종별로 꽃잎길이, 꽃잎폭의 평균을 구하라
def mean_petal(df) :
    return df.mean()
iris.groupby(iris.species).apply(mean_petal)
```

```python
iris.groupby(iris.species).agg(np.mean)
```

```python
iris.groupby(iris.species).agg(np.mean).loc[:,['petal_length','petal_width']]
```

```python
iris.groupby(iris.species).agg(['min','max'])
```

- filter() : 그룹 객체에 filter() 적용할 때 조건식을 가진 함수를 전달하면 조건이 참인 그룹만 남는다

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
class_titanic = titanic.groupby(['class'])
class_titanic
```

```python
# 데이터의 갯수가 200개 이상인 그룹만 필터링 하여라
def filter_func(x) :
    return len(x) >= 200
```

```python
grp_filter = class_titanic.filter(filter_func) 
grp_filter
```

```python
# age 열의 평균이 30보다 작은 그룹만 필터링한다면?
class_titanic.filter(lambda x : x.age.mean() < 30)
```

- pivot(index, columns, value)
- pivot_table(data, values, columns, aggfunc="mean")

```python
# sex, pclass
df = titanic.loc[:,['sex','pclass']]
df
```

```python
df = pd.DataFrame(titanic, columns=['sex','pclass'])
df
```

```python
# 성별과 객실등급에 따른 승객 수 집계하고 싶다면?
# groupby

titanic.groupby(['sex','pclass']).size()
```

```python
df = titanic.groupby(['sex','pclass']).size().reset_index(name='cnt')
df
```

```python
df.pivot('sex','pclass','cnt')
```

```python
# 성별과 생존여부에 따른 승객 수를 집계한다면?
df = titanic.groupby(['sex','survived']).size().reset_index(name='cnt')
df.pivot('sex','survived','cnt')
```

```python
# 성별과 객실등급에 따른 승객 수 집계한다면?
df.pivot_table(index='sex',columns='pclass', values='cnt', aggfunc=np.sum)
```
